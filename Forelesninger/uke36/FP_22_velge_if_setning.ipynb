## Velge type if-setning

Denne notebooken tar for seg hvordan man bestemmer hvilken type if-setninger man vil benytte seg av.

````python
import numpy as np
import matplotlib.pyplot as plt
````

#### Enkel if-setning
Den enklest formen for if-setning er når en operasjon kun skal utføres dersom en spesifikk betingelse er oppfylt og ingenting skal skje ellers. I kodesnutten under plotter vi y = x^2 dersom variabelen skal_plottes == "ja". Dersom ikke gjør koden ingenting.

````python
# Eksempel 1

x = np.linspace(0, 10, 10)
y = x**2

skal_plottes = "ja"

if skal_plottes == "ja":
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
````

#### If else

Den neste typen er if else setninger. Disse bruker vi når vi skal gjøre en handling dersom en betingelse er oppfylt og en annen hvis betingelsen ikke er oppfylt. I koden under plottes det et linjeplot dersom `skal_plottes=="linje"`, ellers plottes det som et scatterplot.

````python
# Eksempel 2

x = np.linspace(0, 10, 10)
y = x**2

skal_plottes = "scatter"

# hvis skal_plottes er lik "linje" skal vi plotte en linje
if skal_plottes == "linje":
    plt.plot(x, y)          # plotter linje
    plt.title("Lineplot")   # setter tittel

# hvis ikke plotter vi et scatterplot    
else:
    plt.scatter(x, y)
    plt.title("Scatterplot")

# Resten er likt for begge tilfeller
plt.xlabel("x") # setter x-akse label
plt.ylabel("y") # setter y-akse label
plt.show()
````

#### If else-if else
If else-if else setninger brukes når vi har flere enn to forskjellige verdi-handlings par. Under kan man spesifisere type plot ved å endre type_plot. Avhenging av verdi vises riktig plot eller så printes det at plottypen er ugyldig.

````python
## Eksempel 3

x = np.linspace(0, 10, 10)
y = x**2

# spesifiser type plot
type_plott = "all" 

# sjekker først om valgt plot er linje
if type_plott == "linje":
    # lineplot med aksetitler og tittel
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Lineplot")
    plt.show()
elif type_plott == "scatter": # sjekker om valgt plot er scatter
    # scatterplot med aksetitler og tittel
    plt.scatter(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Scatterplot")
    plt.show()
elif type_plott == "bar": # sjekker om valgt plot er bar
    # barplot med aksetitler og tittel
    plt.bar(x, y)
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Barplot")
    plt.show()
elif type_plott == "all": # hvis valgt plot er "all" plotter vi alle

    # plt.subplots kommandoen lager en figur med tre akse objekter
    # hvor første parameter er antall rader, andre er antall kolonner
    # og figsize setter størrelsen på figuren
    fig, axs = plt.subplots(1,3, figsize=(18, 6))

    # plt.setp setter delte parametre for alle plotsene
    plt.setp(axs[:], xlabel='x', ylabel='y')

    # plotter linje, scatter og bar i hver sin akse av subplotet
    axs[0].plot(x, y) # lineplot on the first axis
    axs[1].scatter(x, y) # scatterplot on the second axis
    axs[2].bar(x, y) # barplot on the third axis
else:
    # snutten kjøres hvis type_plott ikke er en av de tillatte
    print("Ugyldig type plot")
````

#### Uavhengige (ortogonale) if-setinger og if-else-setninger

I eksemplet over så er det uansett høyst én av betingelsene i if-grenene som vil være sann, så vil ville fått samme resultat om vi hadde hadd separate if-setninger:

```python
if type_plott == "linje":
    (kode)
if type_plott == "scatter":
    (kode)
if type_plott == "bar":
    (kode)
```
Det er tydeligere  å bruke if-else hvis det faktisk bare er en gren/ett alternativ som skal uføres.
 
**Hvis flere av betingelsen kan være sanne, og vi ønsker å utføre kode for hvert tilfelle**, må vi bruke flere uavhengige (ortogonale) if-setninger.

I eksempelet under tenker vi oss at bestilling er en liste med varer som en kunde har bestillt på en restaurant.
Vi må da ha egen if setning for hver vare siden man kan bestille både cola og pizza samtidig. 

````python
# Eksempel 4

# bestillingen kunden har gitt oss
bestilling = ["pizza", "cola", "is", "pizza", "cola"]

# tre variabler som holder styr på hvor mange vi skal ha av hver vare
pizza = 0
cola = 0
ice = 0

# her må vi ha tre if-setninger siden vi kan bestille flere varer
if "pizza" in bestilling:
    pizza = bestilling.count("pizza") # teller hvor mange ganger pizza står i listen
if "cola" in bestilling:
    cola = bestilling.count("cola") # teller hvor mange ganger cola står i listen
if "is" in bestilling:
    ice = bestilling.count("is") # teller hvor mange ganger is står i listen

print("Det skal bestilles %d pizzaer, %d is og %d cola" % (pizza, ice, cola))

````
