```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

En nærmere forklaring på kodelinjene i eksemplet:
- de to første linjene importerer biblioteker
    - __as plt__ og __as np__ lar oss skrive korte prefiks plt og np lenger nede
    - ellers måtte vi ha skrevet hele __matplotlib.pyplot__ og __numpy__ hver gang
- de tre siste kodelinjene lager selve plottet:
    - __fig, ax = plt.subplots( )__ oppretter et figurobjekt fig og et akseobjekt ax
        - figurobjekter kan inneholde ett eller flere akseobjekt
        - akseobjektet er det som inneholder selve plottet (akser, graf)
    - ax.plot(x, y) lager et graf-plott med de gitte x- og y-verdiene
- de to midterste kodelinjene lager x- og y-verdiene som skal plottes
    - __ x = np.linspace(0, 2\*np.pi, 200)__ lager x-verdiene for grafen. De tre argumentene til funksjonen er
        - 0, minste x-verdi vi skal plotte (dvs. plottet starter på x=0)
        - 2$\pi$ (altså ca. 6.28), største x-verdi vi skal plotte
        - 200, antall jevnt fordelte tall som tallserien skal inneholde
    - y = np.sin(x) lager y-verdiene for grafen, ved å regne ut sin(x) for hver x-verdi

Å vise 200 tall på skjermen blir litt mye. Viser resultat av linspace med 10 i stedet:

Vi kan få til de plottene vi var ute etter ved å
- ta utgangspunkt i eksempelkoden
- endre kodelinjen som lager x-verdier til det intervallet vi ønsker
- endre kodelinjen for y-verdier til den funksjonen vi ønsker

For x-verdier ønsket vi intervallet 0-5, altså

```python
x = np.linspace(0, 5, 51)
x
```

Hvis vi skal plotte funksjonen f(x) ønsker vi for y-verdier:

```python
y = f(x)
y
```

Satt sammen til en helhet gir dette, for plotting av f(x)

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2 - 3*x - 1

x = np.linspace(0, 5, 51)
y = f(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

Og for plotting av g(x)

```python
import matplotlib.pyplot as plt
import numpy as np

def g(x):
    return np.sin(np.sqrt(x))

x = np.linspace(0, 5, 51)
y = g(x)

fig, ax = plt.subplots()
ax.plot(x, y)
plt.show()
```

Vi kan også plotte begge funksjonene i samme akseobjekt:

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return x**2 - 3*x - 1

def g(x):
    return np.sin(np.sqrt(x))

x = np.linspace(0, 5, 51)
y1 = f(x)
y2 = g(x)

fig, ax = plt.subplots()
ax.plot(x, y1)
ax.plot(x, y2)
plt.show()
```
