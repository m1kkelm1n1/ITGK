__Frste type feil: Ulovlig variabelnavn / feil tilordning__

Koden under prver  gi verdi til tre variable, gange dem sammen og printe resultatet p skjermen. 

Det er tre feil som vi kan finne og rette:
- 1 = ost er feil rekkeflge i en tilordning
    - Variabelnavnet __ost__ m til venstre, verdien hyre
- __2mat__ er et ulovlig navn. Navn (p variable, funksjoner, ...) 
    - m begynne med _ eller bokstav
    - deretter kan vi ha _ , bokstav eller tall
- __is__ er et ulovlig navn; det er et nkkelord i Python
    - har en spesiell betydning

```python
1 = ost
2mat = 2
is = 3
print(f'Produktet er {ost*2mat*is}')
```

## Feilskrevet nkkelord (keyword)
Nkkelord er reserverte ord med spesiell betydning i Python. I vrt eksempel fins det fire: __import, as, def, return__. 

Nkkelord str med __fet__ grnn skrift i Jupyter Notebook. Andre Python-editorer kan ha andre fargekodinger, men de aller fleste skiller tydelig ut nkkelord.

I eksemplet under har vi skrevet __return__ feil (retur).

```python
def volum_kule(r):
    retur 4.1887902047863905 * r ** 3

radius = float(input('Oppgi radius til kula: '))
volum = round(volum_kule(radius), 2)
print('Volum av kula er', volum)
```

Feilmeldingen peker p tallet 4 selv om feilen er i ordet foran

rsak: ikke grammatisk feil  skrive ordet _retur_ i koden
- kan f.eks. vre et variabelnavn
- men det er feil med _navn mellomrom tall_

NB: Fargekodingen - retur har _ikke_ __fet__ grnn skrift
- ved  flge med p fargene mens du skriver kode, kan du unng en del feil

I eksemplet under har vi glemt kolon bakerst i def-setninga.

## Feil med kommentartegn \# 
Kommentarer blir ikke kjrt, de inneholder forklaringer til koden. I kommentaren (som ogs inni tekststrenger) trenger man derfor ikke flge Python-syntaks, men kan skrive hva som helst, f.eks. vanlig norsk eller engelsk.

Glemt \# foran en slik kommentar vil dermed typisk gi syntaksfeil, da Python prver  tolke det som vanlig kode:

```python
def volum_kule(r):
    4.18... er (4/3)*pi, sparer tid p / og *
    return 4.1887902047863905 * r ** 3

radius = float(input('Oppgi radius til kula: '))
volum = round(volum_kule(radius), 2)
print('Volum av kula er', volum)
```

Nedenfor har vi satt kommentartegn bde p funksjonshodet og frste linje inni funksjonen, mens det bare var linja inni som skulle vre kommentar. 

```python
#def volum_kule(r):
    #4.18... er (4/3)*pi, sparer tid p / og *
    return 4.1887902047863905 * r ** 3

radius = float(input('Oppgi radius til kula: '))
volum = round(volum_kule(radius), 2)
print('Volum av kula er', volum)
```

## Feil / manglende innrykk
Etter setninger som avsluttes med kolon, slik som linje 3 def... i eksemplet vrt, m den eller de neste kodelinjene komme p innrykk. 

Innrykk viser hvilke kodelinjer som er del av funksjonen. Mangel p slikt innrykk vil gi feil.

Som vi ser av feilmeldingen har denne et spesielt navn (IndentationError) men dette er ogs en form for syntaksfeil. 

```python
def volum_kule(r):
'''Tar inn radius r, returnerer volum til kule
    basert p formelen V = (4/3) * pi * r ** 3'''
    # 4.18... er (4/3)*pi, sparer tid p / og *
    return 4.1887902047863905 * r ** 3

radius = float(input('Oppgi radius til kula: '))
volum = round(volum_kule(radius), 2)
print('Volum av kula er', volum)
```

```python
def volum_kule(r):
        '''Tar inn radius r, returnerer volum til kule
    basert p formelen V = (4/3) * pi * r ** 3'''
    # 4.18... er (4/3)*pi, sparer tid p / og *
    return 4.1887902047863905 * r ** 3

radius = float(input('Oppgi radius til kula: '))
volum = round(volum_kule(radius), 2)
print('Volum av kula er', volum)
```

## Oppsummering

Syntaksfeil skyldes brudd p Pythons grammatikk
- typiske rsaker kan vre
    - ulovlige navn
    - feilskrevne nkkelord
    - parenteser eller tekstfnutter som begynnes eller avsluttes feil
    - manglende, eller overfldige, skilletegn
    - manglende, eller overfldige, kommentartegn
    - for f, eller for mange, innrykk

Finne ut av feilene
- feilmeldingen indikerer hvilken kodelinje, og hva i linja
- noen ganger oppdages feilen lenger bak i koden enn den egentlig er
    - lenger bak p samme linje
    - p neste linje
    - eller helt sist i koden (f.eks. uavsluttet trippelfnutt)
    
Se derfor ogs p linjer foran hvis du ikke finner noe galt p indikert linje
