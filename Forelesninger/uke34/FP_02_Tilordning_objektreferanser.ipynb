## Tilordningssetninger: sentrale konsept

- __navnerommet__ inneholder de navnene vi har definert i et program
- navn i navnerommet viser til __objekter__ i minnet
    - objekter kan være f.eks. heltall, flyttall, strenger, lister, funksjoner, ...
- tilordningssetningen knytter et navn til et objekt
    - vi sier gjerne at variabelen har en __referanse__ til objektet

## Eksempelkode: opprette, redefinere, glemme

```python
a = 2.0             
b = 1.5
a =  b + 3
a
```

__Første kodelinje:__ utfører høyre side først, oppretter objekt 2.0
- dvs. et objekt av type float



__Første kodelinje, forts.:__ oppretter navnet a, knytter det til objektet



__Andre kodelinje__: oppretter objekt 1.5, og navn b som knyttes til dette



__Tredje kodelinje__: regner ut b+3 som blir 4.5, oppretter det objektet
    


__Tredje kodelinje, forts:__ navn a fins allerede, redefineres nå til objektet 4.5
- 2.0 blir glemt fordi ingen navn viser til det, minneplass blir frigjort
    


__Fjerning av navn__ fra navnerommet med operatoren __del__
- verdiobjektet 1.5 glemmes også (om ikke andre navn viser til det)
    


```python
del b
b
```

## Hvordan knytte flere navn til samme objekt?

Dette skjer når __høyre__ side er __kun en variabel__

Kort eksempelkode:

```python
a = 2.0
b = a
c = 2.0
d = c + 0.0   
```

### Hva skjer linje  for linje? Starter på første kodelinje
- oppretter objektet 2.0, og navnet a som knyttes til dette



### Andre kodelinje
- her er høyre side __kun__ variabelnavnet a
- opprettet __ikke__ nytt objekt; knytter b til samme objekt som a



### Tredje kodelinje
- med verdien 2.0 på høyre side, blir det derimot laget et nytt objekt
    - selv om dette har samme verdi som et eksisterende objekt
   


### Fjerde kodelinje
- også med et uttrykk som skal regnes ut, blir det et nytt objekt
    - selv når utregningen ikke endrer verdien
    


### Multi-tilordninger

- Flere navn til samme objekt kan også oppnås med én kodelinje
    - = mellom hvert variabelnavn
- Flere tilordninger med komma mellom gir derimot ulike objekter

```python
white_time = black_time = 300.0      # begge variable vil vise til samme objekt
my_points, your_points = 0.0, 0.0    # variablene får samme verdi, men IKKE samme objekt
x, y = 3.2, 1.5                      # variablene får ulike verdier
```

## Hvordan teste for lik verdi og identisk objekt?

Funksjonen id() gir en variabels minneadresse
- hvis id(a) og id(b) har det samme tallet, viser a og b til samme objekt

__==__ og __is__
- testen __a == b__ sjekker om de to variablene har samme verdi
    - merk: dobbelt likhetstegn == for å sjekke likhet
- testen __a is b__ sjekker om de er samme objekt

Hvis resultatet er __True__ er det samme, ellers __False__

```python
a = 2.0
b = a
c = 2.0
d = c + 0.0
# Sjekke verdiene til a, b, c, d
a, b, c, d
```

```python
# Sjekke hvor i minnet a, b, c, d ligger
id(a), id(b), id(c), id(d)
```

```python
# Alternativ sjekk om noe er samme objekt
b is a, c is a, d is a
```

```python
# Operatoren == sjekker om v.s. og h.s. har samme verdi
b == a, c == a, d == a
```

## Spiller det noen rolle om det er samme eller ulikt objekt?
- Samme objekt sparer plass i minnet
    - hvis variablene likevel skulle ha samme verdi
    - nyttig for data som tar stor plass, f.eks. lange lister eller array

__NB:__ Viktig å holde tunga rett i munnen for liste og array
- skal to variable kunne endres uavhengig av hverandre eller ikke?

```python
mine_favorittlag = ["Florø", "Burnley", "Palermo"]
# vennens favorittlag viser til samme listeobjekt
vennens_favorittlag = mine_favorittlag
# Den svikefulle vennen vender det ene favorittlaget ryggen
vennens_favorittlag[0] = "Rosenborg"
vennens_favorittlag, mine_favorittlag
```

vennens\_favorittlag[0] = "Rosenborg" er en såkalt __muterende endring__
- dvs. vi endrer innholdet __inni__ lista heller enn å redefinere til et nytt listeobjekt
- siden vennens\_favorittlag og mine\_favorittlag refererer til samme objekt gjør dette at begge variablene påvirkes av denne endringen av innhold

I eksemplet under skjer derimot ikke dette:

```python
mine_favorittlag = ["Florø", "Burnley", "Palermo"]
# vennens_favorittlag viser til samme listeobjekt
vennens_favorittlag = mine_favorittlag
# Den svikefulle vennen vender det ene favorittlaget ryggen
vennens_favorittlag = ["Rosenborg", "Burnley", "Palermo"]
vennens_favorittlag, mine_favorittlag
```

vennens\_favorittlag = \["Rosenborg", "Burnley", "Palermo"\] er derimot en ikke-muterende endring
- variabelen vennens\_favorittlag redefineres til et nytt listeobjekt
- dette påvirker dermed ikke variabelen mine\_favorittlag som fortsetter å vise til den gamle lista

### En bedre måte å gjøre det på
siden mine og vennens favorittlag bør kunne endres uavhengig av hverandre:
- gjør dem til separate listeobjekt allerede når vi oppretter variabelnavnene
- da vil aldri endring i vennens\_favorittlag implisitt også endre mine

```python
mine_favorittlag = ["Florø", "Burnley", "Palermo"]
# vennens_favorittlag blir nå et separat listeobjekt:
vennens_favorittlag = ["Florø", "Burnley", "Palermo"]
# Den svikefulle vennen endrer et favorittlag
vennens_favorittlag[0] = "Rosenborg"
vennens_favorittlag, mine_favorittlag
```

For immuterende datatyper som __float__ er det ingen fare for tabbe her:

```python
white_time = black_time = 300.0
white_time[0] = white_time - 4.25
white_time, black_time
```

## Oppsummering
Tilordningssetninger har syntaks

\< navn \> = \< uttrykk \>
- regner ut uttrykket
- oppretter objekt i minnet (for resultat av høyre side)
    - unntatt h.s. __kun__ et navn, da gjenbrukes gammelt objekt
- oppretter navnet i navnerommet
    - om ikke navnet fins allerede (da redefineres variabelen)
- knytter navnet til objektet
    - evt. gammelt objekt som ingen navn viser til, glemmes

Flere navn kan vise til samme objekt
- nyttig, sparer minneplass, særlig for lister, array, ...
    - viktig å holde tunga rett i munnen ved muterende endringer
    - unngå utilsiktet endring av én variabel via en annen
