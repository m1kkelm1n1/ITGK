## Variable: Huske data mens programmet kjører
__Uten__ variable:
- må regne ut på nytt tall vi har regnet ut før
    - sløser regnetid og strøm
- spørre brukeren om info vi har spurt om før
    - sløser tid for brukeren, irriterende

Eksemplet under viser et program hvor vi regner ut noen data, men
- __ikke__ husker noen av dataene i programmet
- bare printer resultatene på skjermen

Utregningen av volum for ei kule, (4/3)\*3.14\*2.5\*\*3, går igjen tre steder
- unødig mange multiplikasjoner, sløser tid
- ikke farlig for dette bitte lille programmet
- men veldig dumt i et større (og nyttigere) program

```python
print('Volum av kule med radius 2.5 cm')
print((4/3)*3.14*2.5**3)
print('Kubikkcentimeter materiale som trengs for 85 kuler')
print(85 * (4/3)*3.14*2.5**3)
print('Antall kuler vi kan produsere gitt 10000 ccm materiale, antatt 10% svinn')
print(10000 // ((4/3)*3.14*2.5**3 * 0.9))
```

I tilsvarende eksempel under, husker vi volumet i en variabel, v_kule

Slipper da unna med å regne det ut én gang

```python
print('Volum av perfekt kule med radius 2.5 cm')
v_kule = (4/3)*3.14*2.5**3
print(v_kule)
print('Kubikkcentimeter materiale som trengs for 85 kuler')
print(85 * v_kule)
print('Antall kuler vi kan produsere gitt 10000 ccm materiale')
print(10000 // v_kule)
```

__Hvordan fungerer tilordningsoperatoren?__

__=__ er et symbol du har sett ofte før, f.eks.
- i matematikk
- i regneark

Men det er store forskjeller på hva den betyr
- matematikk: påstand, "er lik"
    - Python: instruks, "gi variabelen verdien ..."
- regneark: varig relasjon mellom innholdet i celler
    - Python: utregning som gjøres i øyeblikket

__= matte: venstre side _er lik_ høyre side__

- Påstand (fortellende setning)
- Eksempel: $(a+b)^2=a^2+2ab+b^2$
- Symmetrisk: $c=d \Leftrightarrow d=c$

__= Python: variabel v.s. _settes lik_ verdi h.s.__
- Instruksjon (imperativ setning): 
    - __x=1__ "Gi x verdien 1"
    - __ikke__ symmetrisk
- v.s. kan __kun__ være en variabel
- h.s. kan være verdi, variabel, sammensatt uttrykk
    - evt. variable på h.s. __må__ være definert tidligere

```python
# Akseptabelt i Python, v.s. variabelnavn, h.s. en verdi
a = 1
b = 2
b = a 
# Sjekke verdier a og b har nå
a,  b 
```

```python
a = 5      # Gir a verdien 5
a = a + 1  # Meningsløst i matte, ok i Python
# a på h.s.: gammel verdi, a på v.s.: ny verdi for a
a
```

__= i regneark vs. Python__

Begge er "imperativsetninger", men viktig forskjell.

Virkemåten til formler i regneark:



__Et Python-program virker annerledes:__

```python
a = 2.5
b = 3.25
c = a + b
b = 1.4
c = a + b
c , b
```

Syntaksregler for tilordningssetninger

\<variabel\> = \<uttrykk\>

Venstre side __må__ være et variabelnavn

- variabelnavn
    - må starte med _ eller bokstav
    - fortsette med _ , bokstav eller tall
    - case-sensitive

Høyre side kan være 
- verdi, variabel eller sammensatt uttrykk
- variable h.s. må ha fått verdi __før__ tilordningssetninga kjøres

```python
#2rabatt = 0.8                         # ikke ok, variabel begynner med tall
rabatt2 = 0.8                          # ok
#pris inkl moms = 14.99                # ikke ok, mellomrom
pris_inkl_moms = 14.99                 # ok
#salgspris = Pris_inkl_moms * rabatt2  # ikke ok, Pris_... udefinert
salgspris = pris_inkl_moms * rabatt2   # ok
# Sjekke verdi for salgspris
salgspris
```

## Oppsummering

Tilordningssetning: \<variabelnavn\> = \<uttrykk\>
- knytter variabelnavnet på venstre side
- til verdien som blir resultat av høyre side

Imperativ setning (ordre), ikke påstand (som = i matematikk)
- x = 5  ("Gi x verdien 5")

Variabelen husker bare __resultatet__, ikke uttrykket på høyre side
- ulikt regneark, hvor cella husker formelen

Venstre side kan __kun__ ha et variabelnavn
- begynne med \_ eller bokstav
- fortsette med \_ , bokstav og tall

Høyre side kan være et større uttrykk
- inkl aritmetiske operatorer, funksjonskall
- evt. variable på høyre side __må__ ha fått verdi før i programmet
