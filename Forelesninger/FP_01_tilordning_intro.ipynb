## Variable: Huske data mens programmet kjrer
__Uten__ variable:
- m regne ut p nytt tall vi har regnet ut fr
    - slser regnetid og strm
- sprre brukeren om info vi har spurt om fr
    - slser tid for brukeren, irriterende

Eksemplet under viser et program hvor vi regner ut noen data, men
- __ikke__ husker noen av dataene i programmet
- bare printer resultatene p skjermen

Utregningen av volum for ei kule, (4/3)\*3.14\*2.5\*\*3, gr igjen tre steder
- undig mange multiplikasjoner, slser tid
- ikke farlig for dette bitte lille programmet
- men veldig dumt i et strre (og nyttigere) program

```python
print('Volum av kule med radius 2.5 cm')
print((4/3)*3.14*2.5**3)
print('Kubikkcentimeter materiale som trengs for 85 kuler')
print(85 * (4/3)*3.14*2.5**3)
print('Antall kuler vi kan produsere gitt 10000 ccm materiale, antatt 10% svinn')
print(10000 // ((4/3)*3.14*2.5**3 * 0.9))
```

I tilsvarende eksempel under, husker vi volumet i en variabel, v_kule

Slipper da unna med  regne det ut n gang

```python
print('Volum av perfekt kule med radius 2.5 cm')
v_kule = (4/3)*3.14*2.5**3
print(v_kule)
print('Kubikkcentimeter materiale som trengs for 85 kuler')
print(85 * v_kule)
print('Antall kuler vi kan produsere gitt 10000 ccm materiale')
print(10000 // v_kule)
```

__Hvordan fungerer tilordningsoperatoren?__

__=__ er et symbol du har sett ofte fr, f.eks.
- i matematikk
- i regneark

Men det er store forskjeller p hva den betyr
- matematikk: pstand, "er lik"
    - Python: instruks, "gi variabelen verdien ..."
- regneark: varig relasjon mellom innholdet i celler
    - Python: utregning som gjres i yeblikket

__= matte: venstre side _er lik_ hyre side__

- Pstand (fortellende setning)
- Eksempel: $(a+b)^2=a^2+2ab+b^2$
- Symmetrisk: $c=d \Leftrightarrow d=c$

__= Python: variabel v.s. _settes lik_ verdi h.s.__
- Instruksjon (imperativ setning): 
    - __x=1__ "Gi x verdien 1"
    - __ikke__ symmetrisk
- v.s. kan __kun__ vre en variabel
- h.s. kan vre verdi, variabel, sammensatt uttrykk
    - evt. variable p h.s. __m__ vre definert tidligere

```python
# Akseptabelt i Python, v.s. variabelnavn, h.s. en verdi
a = 1
b = 2
b = a 
# Sjekke verdier a og b har n
a,  b 
```

```python
a = 5      # Gir a verdien 5
a = a + 1  # Meningslst i matte, ok i Python
# a p h.s.: gammel verdi, a p v.s.: ny verdi for a
a
```

Syntaksregler for tilordningssetninger

\<variabel\> = \<uttrykk\>

Venstre side __m__ vre et variabelnavn

- variabelnavn
    - m starte med _ eller bokstav
    - fortsette med _ , bokstav eller tall
    - case-sensitive

Hyre side kan vre 
- verdi, variabel eller sammensatt uttrykk
- variable h.s. m ha ftt verdi __fr__ tilordningssetninga kjres

```python
#2rabatt = 0.8                         # ikke ok, variabel begynner med tall
rabatt2 = 0.8                          # ok
#pris inkl moms = 14.99                # ikke ok, mellomrom
pris_inkl_moms = 14.99                 # ok
#salgspris = Pris_inkl_moms * rabatt2  # ikke ok, Pris_... udefinert
salgspris = pris_inkl_moms * rabatt2   # ok
# Sjekke verdi for salgspris
salgspris
```

## Oppsummering

Tilordningssetning: \<variabelnavn\> = \<uttrykk\>
- knytter variabelnavnet p venstre side
- til verdien som blir resultat av hyre side

Imperativ setning (ordre), ikke pstand (som = i matematikk)
- x = 5  ("Gi x verdien 5")

Variabelen husker bare __resultatet__, ikke uttrykket p hyre side
- ulikt regneark, hvor cella husker formelen

Venstre side kan __kun__ ha et variabelnavn
- begynne med \_ eller bokstav
- fortsette med \_ , bokstav og tall

Hyre side kan vre et strre uttrykk
- inkl aritmetiske operatorer, funksjonskall
- evt. variable p hyre side __m__ ha ftt verdi fr i programmet
